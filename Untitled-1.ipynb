```python
# Function to filter data by state
def filter_data_by_state(state):
    return data[data['State'] == state]

# Function to aggregate data for low zoom level
def aggregate_data():
    return data.groupby('State').size()

# Example usage
start_time = time.time()
state_data = filter_data_by_state('CA')
agg_data = aggregate_data()
processing_time = time.time() - start_time

print("Aggregated Data:", agg_data)
print("Processing Time:", processing_time)
```
